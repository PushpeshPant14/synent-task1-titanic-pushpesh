```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Age'].isnull().sum()
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

Converting sex to numerical value
male - 0
female - 1

```python
df['Sex'].unique()
```

```python
df['Sex']=df['Sex'].map({'male':0,'female':1})
```

```python
df['Sex'].dtype
```

```python
df.head()
```

```python
df['Embarked'].unique()
```

```python
df = pd.get_dummies(df, columns=['Embarked'])
```

```python
df.head()
```

```python
df.rename(columns={
    'PassengerId': 'passenger_id',
    'Survived': 'survived',
    'Pclass': 'passenger_class',
    'Name': 'name',
    'Sex': 'sex',
    'Age': 'age',
    'SibSp': 'siblings_spouse',
    'Parch': 'parents_children',
    'Fare': 'fare',
}, inplace=True)
```

```python
print(df.info())
print(df.head())
print(df.isnull().sum())
```

```python
df.to_csv("clean_titanic.csv", index=False)
```
